```python
Initialization: Select ( k ) initial centroids randomly from the dataset.

Assignment: Assign each data point to the nearest centroid, forming ( k ) cluster

.
Update: Calculate the new centroids by taking the mean of all data points in each cluse

r.
Iteration: Repeat the assignment and update steps until the centroids no longer change significantly or a maximum number of iterations is reached.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)

data ={
    "Age" : np.random.randint(18, 70, size=100),
    "Income": np.random.randint(2000, 100000, size=100),
    "Spending_Score": np.random.randint(1, 100, size=100)
}

df =pd.DataFrame(data)
df.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="Age", y="Income", hue="Spending_Score", palette="viridis", data=df)
plt.title("Employes Datasets")
plt.show()
```

# Implementing K-Means using sklearn 

```python
from sklearn.cluster import KMeans
```

```python
Kmeans =KMeans(n_clusters=3, random_state=1)
Kmeans.fit(df)
```

```python
df["Cluster"] = Kmeans.labels_
```

```python
df.head()
```

```python
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x="Age", y="Income", hue="Cluster", palette="viridis")
plt.title("Employee of Year")
plt.show()
```
